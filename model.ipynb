```python
class LabeledLineSentence(object):
    def __init__(self, sources):
        self.sources = sources
        
        flipped = {}
        
        # make sure that keys are unique
        for key, value in sources.items():
            if value not in flipped:
                flipped[value] = [key]
            else:
                raise Exception('Non-unique prefix encountered')
    
    def __iter__(self):
        for source, prefix in self.sources.items():
            with utils.smart_open(source) as fin:
                for item_no, line in enumerate(fin):
                    yield LabeledSentence(utils.to_unicode(line).split(), [prefix + '_%s' % item_no])
    
    def to_array(self):
        self.sentences = []
        for source, prefix in self.sources.items():
            with utils.smart_open(source) as fin:
                for item_no, line in enumerate(fin):
                    self.sentences.append(LabeledSentence(utils.to_unicode(line).split(), [prefix + '_%s' % item_no]))
        return self.sentences
    
    def sentences_perm(self):
        shuffle(self.sentences)
        return self.sentences
```

```python
# asi - 100
# minpromtorgrus - 010
# robolaw - 001

DATA = []
DATA_CAT = []
```

```python
for t, cs in task:
    for j in list(getallwall(domain=t, mx=1000)):
        
        try:
            w = clear(j['text'])
            if len(w) == 0: continue
            DATA.append(w)
            DATA_CAT.append(cs)
        except IndexError:
            pass
```

```python
old = -1 
BASE_WORD = ['аси'] #такие не будем приводить к начальной форме

for i in range(len(DATA)):
    nr = round(100*i/len(DATA))
    if old != nr: 
        old = nr
        print('$', nr)
    j = DATA[i].lower().split()
    
    for kk in range(len(j)):
        if j[kk] in BASE_WORD: continue
        j[kk] = morph.parse(j[kk])[0].normal_form #машин -> машина
        
    DATA[i] = j
```

```python
from collections import Counter

vocab = Counter()
```

```python
for text in DATA:
    for word in text:
        vocab[word]+=1
```

```python
total_words = len(vocab)

def get_word_2_index(vocab):
    word2index = {}
    for i,word in enumerate(vocab):
        word2index[word.lower()] = i
        
    return word2index

word2index = get_word_2_index(vocab)
#фактически на выходе получается вектор, где на каждой позиции встречается определенное слово. очевидно, от этой системы хранения информации нужно уйти
```

```python
def get_batch(X, Y,i,batch_size):
    batches = []
    results = []
    texts = X[i*batch_size:i*batch_size+batch_size]
    categories = Y[i*batch_size:i*batch_size+batch_size]
    for text in texts:
        layer = np.zeros(total_words,dtype=float)
        for word in text:
            layer[word2index[word]] += 1
            
        batches.append(layer)
        
    for category in categories:
        y = np.zeros((TOTAL_CAT),dtype=float)
        y[category] = 1.
        results.append(y)
            
    return np.array(batches),np.array(results)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(DATA, DATA_CAT, test_size=0.1, random_state=54)
```

```python
# Parameters
learning_rate = 0.01
training_epochs = 10
batch_size = 150
display_step = 1

# Network Parameters
n_hidden_1 = 100      # 1st layer number of features
n_hidden_2 = 100       # 2nd layer number of features
n_input = total_words # Words in vocab
n_classes = TOTAL_CAT         # Categories: graphics, sci.space and baseball

input_tensor = tf.placeholder(tf.float32,[None, n_input],name="input")
output_tensor = tf.placeholder(tf.float32,[None, n_classes],name="output") 
```

```python
def multilayer_perceptron(input_tensor, weights, biases):
    layer_1_multiplication = tf.matmul(input_tensor, weights['h1'])
    layer_1_addition = tf.add(layer_1_multiplication, biases['b1'])
    layer_1 = tf.nn.relu(layer_1_addition)
    
    # Hidden layer with RELU activation
    layer_2_multiplication = tf.matmul(layer_1, weights['h2'])
    layer_2_addition = tf.add(layer_2_multiplication, biases['b2'])
    layer_2 = tf.nn.relu(layer_2_addition)
    
    # Output layer 
    out_layer_multiplication = tf.matmul(layer_2, weights['out'])
    out_layer_addition = out_layer_multiplication + biases['out']
    
    return out_layer_addition
```

```python
# Store layers weight & bias
weights = {
    'h1': tf.Variable(tf.random_normal([n_input, n_hidden_1])),
    'h2': tf.Variable(tf.random_normal([n_hidden_1, n_hidden_2])),
    'out': tf.Variable(tf.random_normal([n_hidden_2, n_classes]))
}
biases = {
    'b1': tf.Variable(tf.random_normal([n_hidden_1])),
    'b2': tf.Variable(tf.random_normal([n_hidden_2])),
    'out': tf.Variable(tf.random_normal([n_classes]))
}

# Construct model
prediction = multilayer_perceptron(input_tensor, weights, biases)

# Define loss and optimizer
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=prediction, labels=output_tensor))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)

# Initializing the variables
init = tf.global_variables_initializer()
```

```python
sess = tf.Session()
sess.run(init)

# Training cycle
for epoch in range(training_epochs):
    avg_cost = 0.
    total_batch = int(len(X_train)/batch_size)
    # Loop over all batches
    for i in range(total_batch):
        batch_x,batch_y = get_batch(X_train, y_train, i,batch_size)
        # Run optimization op (backprop) and cost op (to get loss value)
        c,_ = sess.run([loss,optimizer], feed_dict={input_tensor: batch_x,output_tensor:batch_y})
        # Compute average loss
        avg_cost += c / total_batch
    # Display logs per epoch step
    if epoch % display_step == 0:
        print("Epoch:", '%04d' % (epoch+1), "loss=", \
            "{:.9f}".format(avg_cost))
print("Optimization Finished!")
```

```python

# Test model
correct_prediction = tf.equal(tf.argmax(prediction, 1), tf.argmax(output_tensor, 1))
# Calculate accuracy
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
total_test_data = len(X_test)
batch_x_test,batch_y_test = get_batch(X_test, y_test, 0, total_test_data)
print("Accuracy:", accuracy.eval({input_tensor: batch_x_test, output_tensor: batch_y_test}, session=sess))
```
